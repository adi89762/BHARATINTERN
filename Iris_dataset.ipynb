```python
Data = pd.DataFrame({'sepal length': iris['SepalLengthCm'],
                     'sepal width': iris['SepalWidthCm'],
                     'petal length': iris['PetalLengthCm'],
                     'petal width': iris['PetalWidthCm'],
                     'species': iris['Species']})
print(Data.head())
```

```python
X = Data[['sepal length','sepal length','sepal width','petal length','petal width']]
```

```python
y = Data['species']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn. ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
X_test
```

```python
clf.score(X_test,y_test)
```
